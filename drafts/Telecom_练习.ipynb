```python
def tenure_lab(df):
    if df['tenure'] <=12:
        return "tenure_0_12"
    elif (df['tenure'] >12) & (df['tenure']<=24):
        return 'tenure_12_24'
    elif (df['tenure']>24) & (df['tenure']<=48):
        return 'tenure_24_48'
    elif (df['tenure']>48) & (df['tenure']<=60):
        return 'tenure_48_60'
    else:
        return "tenure_gt_60"
```

```python
df["tenure_group"]=df.apply(lambda df:tenure_lab(df),axis=1)
# 这里的这个   axis=1   有点出乎意料。
```
